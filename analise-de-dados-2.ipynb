```python
import pandas as pd

caminho_planilha = 'C:/Users/55469/Downloads/analise_de_dados_ativ.xls'

df = pd.read_excel(caminho_planilha)
```

```python
bill_columns = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']

pay_columns = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
```

```python
bill_summary = df[bill_columns].describe()
print(bill_summary)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))  
fig.suptitle('Histogramas dos Valores das Faturas')

for i, column in enumerate(bill_columns):
    ax = axes[i // 3, i % 3] 
    df[column].plot(kind='hist', bins=20, ax=ax, title=column)
    ax.set_xlabel('Valor')
    ax.set_ylabel('Frequncia')

plt.tight_layout()
plt.show()
```

```python
pay_summary = df[pay_columns].describe()
print(pay_summary)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))  
fig.suptitle('Histogramas dos Valores de Pagamento')

for i, column in enumerate(pay_columns):
    ax = axes[i // 3, i % 3]  
    df[column].plot(kind='hist', bins=20, ax=ax, title=column)
    ax.set_xlabel('Valor')
    ax.set_ylabel('Frequncia')
    ax.tick_params(axis='x', rotation=45) 

plt.tight_layout()
plt.show()
```

```python
pay_zero_mask = df[pay_columns] == 0

pay_zero_counts = pay_zero_mask.sum()
print(pay_zero_counts)
```

```python
import numpy as np
import matplotlib.pyplot as plt

pay_non_zero = df[pay_columns][df[pay_columns] != 0]
pay_log = pay_non_zero.apply(np.log10)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))  
fig.suptitle('Histogramas das Transformaes Logartmicas dos Pagamentos')

for i, column in enumerate(pay_columns):
    ax = axes[i // 3, i % 3]  
    pay_log[column].plot(kind='hist', bins=20, ax=ax, title=column)
    ax.set_xlabel('Log10(Valor)')
    ax.set_ylabel('Frequncia')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
